```python
import zipfile
from PIL import Image
import os
import pickle

zip_file_path = 'nyu2.zip'

extracted_images_dir = 'extracted_images'
os.makedirs(extracted_images_dir, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as z:
    z.extractall(extracted_images_dir)
```

```python
root = os.listdir(extracted_images_dir)[0]
print(os.path.join(extracted_images_dir, root))
print(os.listdir(os.path.join(extracted_images_dir, root, root)))
```

```python
import numpy as np
from torchvision import transforms

class Resize(object):
    def __init__(self, output_resolution):
        self.resize = transforms.Resize(output_resolution)
    
    def __call__(self, sample):
        image, depth = sample['image'], sample['depth']

        if isinstance(image, np.ndarray):
            image = Image.fromarray(np.uint8(image))
        if isinstance(depth, np.ndarray):
            depth = Image.fromarray(depth)
        
        image = self.resize(image)
        depth = self.resize(depth)

        return {'image': image, 'depth': depth}
```

```python
image_data_pairs = []
resize_transform = Resize((240, 320))
for png_file in os.listdir(os.path.join(extracted_images_dir, root, root)):
    if png_file.endswith('.png'):
        jpg_file = png_file.replace('.png', '.jpg')
        if jpg_file in os.listdir(os.path.join(extracted_images_dir, root, root)):
            png_img = Image.open(os.path.join(extracted_images_dir, root, root, png_file))
            jpg_img = Image.open(os.path.join(extracted_images_dir, root, root, jpg_file))
            
            sample = resize_transform({'image': jpg_img, 'depth': png_img})

            image_data_pairs.append((sample['image'], sample['depth']))
```

```python
print(image_data_pairs)

pickle_file_path = 'nyu_buffer.pickle'
with open(pickle_file_path, 'wb') as f:
    pickle.dump(image_data_pairs, f)

print('Success')
```
